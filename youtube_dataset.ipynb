```python
data["category"].replace(np.nan,'Entertainment',inplace=True)
data
```

```python
data["Country"].replace(np.nan,'United States',inplace=True)
data
```

```python
data["channel_type"].replace(np.nan,'Entertainment',inplace=True)
data
```

```python
avg_country_rank =data["country_rank"].astype("float").mean(axis=0)
data["country_rank"].replace(np.nan,avg_country_rank,inplace=True)
```

```python
avg_video_views_rank =data["video_views_rank"].astype("float").mean(axis=0)
data["video_views_rank"].replace(np.nan,avg_video_views_rank,inplace=True)
```

```python
avg_channel_type_rank  =data["channel_type_rank"].astype("float").mean(axis=0)
data["channel_type_rank"].replace(np.nan,avg_channel_type_rank,inplace=True)
```

```python
avg_video_views_for_the_last_30_days  =data["video_views_for_the_last_30_days"].astype("float").mean(axis=0)
data["video_views_for_the_last_30_days"].replace(np.nan,avg_video_views_for_the_last_30_days,inplace=True)
```

```python
avg_subscribers_for_last_30_days =data["subscribers_for_last_30_days"].astype("float").mean(axis=0)
data["subscribers_for_last_30_days"].replace(np.nan,avg_subscribers_for_last_30_days,inplace=True)
```

```python
data["created_year"].replace(np.nan,'2014.0',inplace=True)
```

```python
data["created_month"].replace(np.nan,'JAN',inplace=True)
```

```python
data["created_date"].replace(np.nan,'19',inplace=True)
```

```python
data["Gross tertiary education enrollment (%)"].replace(np.nan,'28.1',inplace=True)
```

```python
avg_Population =data["Population"].astype("float").mean(axis=0)
data["Population"].replace(np.nan,avg_Population,inplace=True)
```

```python
data["Unemployment rate"].replace(np.nan,'14.7',inplace=True)
```

```python
avg_Urban_population =data["Urban_population"].astype("float").mean(axis=0)
data["Urban_population"].replace(np.nan,avg_Urban_population,inplace=True)
```

```python
avg_Latitude =data["Latitude"].astype("float").mean(axis=0)
data["Latitude"].replace(np.nan,avg_Latitude,inplace=True)
```

```python
avg_Longitude =data["Longitude"].astype("float").mean(axis=0)
data["Longitude"].replace(np.nan,avg_Longitude,inplace=True)
```

```python
data["Abbreviation"].replace(np.nan,'BD',inplace=True)
```
